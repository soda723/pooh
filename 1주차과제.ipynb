## **1. 언어 분야**


언어 분야의 인공지능 하면 자연스럽게 가장 먼저 떠오르는 것은 번역기이다. 그 중에서도 한국에서 많이 사용하는 번역기는 네이버 파파고와 구글 번역이라고 생각한다. 그 중 네이버 **파파고**에 대해서 조사해 보았다.

네이버 파파고 홈페이지에 들어가 하단 부분을 살펴보면 공개 API 라는 항목이 있다. 그 항목을 통해 이동하면 공개 API를 지원받을수 있는 홈페이지와 함께 간략한 설명도 찾아볼 수 있다. ([네이버 파파고 공개 API 지원 주소](https://developers.naver.com/products/nmt/))

![파파고](https://user-images.githubusercontent.com/60220571/83666784-728cca00-a608-11ea-8f99-3fd7095766f7.jpg)

홈페이지 설명을 살펴보면 **파파고의 NMT 기술**은 입력 문장을 문장벡터로 변환하는 신경망(encoder)와 문장 벡터에서 번역하는 언어의 문장을 생성하는 신경망(decoder)를 대규모의 병렬 코퍼스부터 자동으로 학습한다고 한다. 또한 문장 전체 정보를 바탕으로 번역을 수행하기 때문에 기존 SMT 방식의 번역보다 더욱 정확하고 문장 맥락에 맞는 번역을 한다.
이 설명을 통해 과거에는 SMT기술을 사용했으나 현재는 NMT 기술을 통해 번역기를 구현하고 있는 것을 알 수 있다. SMT기술과 NMT 기술을 살펴보면, 

먼저 **SMT**은 Statistical Machine Translation의 약자로 **통계 기반 번역**이라는 의미이다. 대규모 데이터를 기반으로 자동으로 학습된 통계적 모델을 통해 번역한다. 단위(구(Phrase)또는 단어)마다 번역을 한 후 이 번역된 단어들을 조합한다. 이 경우 한국어와 일본어 같이 어순이 비슷한 경우 조금 더 낫지만 한국어와 영어처럼 어순 배열이 다른 경우 올바른 배열을 찾기 어려워 부자연스럽다고 느낄수 있다. 또한 동음이의어의 경우 예를 들어, 배를 번역할때, ship과 pear 등 2가지 이상의 단어로 번역된다는 등의 오류 발생 가능성이 높다고 할 수 있다. 참고로 SMT 버전의 파파고 API 제공은 2019년 11월 30일에 종료되었다고 한다.

**NMT**는 Neural Machine Tanslation의 약자로 **인공신경망 기반 기계 번역**이라는 의미이다. NMT는 Attention mechanism(어텐션 메커니즘)을 바탕으로 한다. 어텐션 메커니즘이란 특정 벡터에 집중해 성능을 최대한 높이는 방법이다. NMT 알고리즘은 문장을 통째로 코드화(인코딩)하고 그 코드를 디코딩하여 번역문을 생성한다. 이 번역문을 올바르게 번역된 문장과 비교하여 오답과 차이를 비교한다. 이후 오답과의 차이를 줄이는 방향으로 인코딩과 디코딩을 반복하여 정답과 가까워지고 이 과정 속에서 스스로 규칙을 학습한다.

현재는 대부분의 번역기들이 NMT 알고리즘을 바탕으로 하고 있어, 예전보다 훨씬 매끄러운 번역문을 만날 수 있다.

## **2. 음성 분야**

음성 분야에서 활용되는 인공지능 제품으로는 대표적으로 인공지능 스피커가 있다. 인공지능 스피커(스마트 스피커, smart speaker)는 무선 스피커의 일종으로서, 하나 이상의 인상적인 말(hot word)의 도움을 받아 상호작용 동작과 핸즈프리 활성화를 제공하는 가상비서가 내장된 **‘보이스 커맨드 디바이스’**이다. 한국에서는 카카오의 카카오미니, 네이버의 클로바, SKT의 NUGU, KT의 기가지니 등의 제품들이 있고 해외에서는 아마존의 아마존 에코, 구글의 구글 홈 등이 있다. 이들 중 **아마존의 에코에 사용된 AI 알렉사**를 바탕으로 인공지능 스피커의 원리에 대해 조사해 보았다.

![아마존 에코](https://user-images.githubusercontent.com/60220571/83725366-c76a2800-a67c-11ea-9654-3a3471c81a6d.jpg)

알렉사가 음성을 인식하는 원리는, 인간의 목소리(명령)가 스피커의 마이크를 통해 입력되면 ‘스피치 프로세서(SPEECH PROCESSOR)’는 음성신호를 음성 신호 벡터로 변화시킨다. 이 음성 신호 벡터는 ‘스피치 인식기(SPEECH RECOGNIZER)’와 ‘언어 특징 추출기(LINGUISTIC REATURE EXTRACTOR)’를 통해 ‘언어 표현 정보(LINGUISTIC REPRESENTATION)’로 추출된다.

추출된 언어 표현 정보는 ‘스킬 분류기(SKILL CLASSIFIER)’, ‘의도분류기(INTENT CLASSIFIER)’, ‘슬롯 인지기(SLOT RECOGINZER)’로 입력된다. 스킬 분류기는 언어 표현 정보를 통해 입력된 명령에 알맞은 스킬을 결정한다. 의도분류기는 언어 표현 정보를 통해 입력된 명령의 의도를 결정한다. 슬롯 인지기는 언어표현 정보에서 의미 있는 정보를 인지한다. 분류 및 인지를 마친 스킬, 의도, 슬롯은 ‘스킬 프로세서’를 거쳐 명령을 위한 관련 기기를 동작시킨다. 

![알렉사](https://user-images.githubusercontent.com/60220571/83725310-aefa0d80-a67c-11ea-8502-b30757210e83.png)

알렉사가 입력된 언어를 제대로 이해하기 위해서는 스킬, 의도, 슬롯에 대한 정확한 분석이 선행되어야 한다. 그렇기 때문에 알렉사의 의도 분류기는 입력된 데이터를 기반으로 학습이 가능하다. **‘인공신경망(Artificial Neural Network, ANN)’와 ‘심층신경망(Deep Neural Network, DNN)’**를 사용할 수 있는 의도 분류기는 축적되는 데이터에 따라 뉴럴 네트워크의 변화하는 가중치를 감지하고, 이를 학습해 더욱 정확하게 명령의 의도를 판단할 수 있게 된다고 한다.

여기서 **인공신경망(ANN)**은 기계학습과 인지과학에서 생물학의 신경망(동물의 중추신경계 중 특히 뇌)에서 영감을 얻은 통계학적 학습 알고리즘으로, 시냅스의 결합으로 네트워크를 형성한 인공 뉴런(노드)이 학습을 통해 시냅스의 결합 세기를 변화 시켜, 문제 해결 능력을 갖추는 모델 전반을 가리킨다. 또 **심층신경망(DNN)**은 입력층(input layer)과 출력층(output layer)상이에 여러 개의 은닉층(hidden layer)으로 이뤄진 인공신경망이다.

인공지능 스피커는 음악이나 TV 등의 콘텐츠뿐만 아니라 IoT를 직접 제어하는 등 앞으로 생활이나 사회 곳곳에 스며들 것이기 때문에 사용자와 아주 밀접하게 존재하여 편의에 도움을 줄 것이다.

## **3. 이미지 분야**

이미지 분야의 경우 다른 분야보다 상대적으로 이미지 분류나, 손 제스처 인식 등 익숙한 기술들이 많았다. 그 중에서 요새 코로나-19로 인해 사람들이 마스크를 착용하고 잠금인식을 해제하려는 경우가 많아지면서 아이폰이 마스크 쓴 얼굴로도 얼굴인식으로 잠금화면을 해제할 수 있는가에 대한 기사를 보고 얼굴인식 분야에 대해 조사해보았다. ([한겨례 - 애플, 아이폰 얼굴인식 기능 개선…마스크 쓰면 곧장 암호 물어]( http://www.hani.co.kr/arti/economy/it/945850.html))

![face id](https://user-images.githubusercontent.com/60220571/83725504-01d3c500-a67d-11ea-993b-a83ff71d4d54.jpg)

**얼굴인식 시스템(안면 인식 시스템, facial recognition system)**은 디지털 이미지를 통해 각 사람을 자동으로 식별하는 컴퓨터 지원 응용 프로그램을 말한다. 이미 아이폰과 갤럭시 시리즈 등 많은 스마트 폰에서 생체인식으로 활용되고 있기 때문에 우리에게 매우 익숙한 기술이기도 하다. 

얼굴인식 시스템의 원리는 **미리 세팅해놓은 환경에서 촬영된 이미지와 데이터베이스에 저장된 검증된 이미지와 대조하는 방식이다.** 얼굴인식 과정은 크게 두 단계를 걸친다. 1단계는 얼굴 부분의 인식이고, 2단계는 그 얼굴이 누구의 얼굴인지 감별하거나 등록된 얼굴이 일치하는지 검사하는 것이다. 

애플의 페이스 ID를 통해 원리를 살펴보면(아이폰X 기준), 기기 전면의 트루뎁스 카메라 시스템을 이용한다. 이 시스템은 적외선 카메라, 투광 일루미네이터, 근접센서, 도트 프로젝터로 이루어져 있다. 트루뎁스 카메라 시스템은 사용자 얼굴을 30,000개 이상의 점으로 입체적으로 스캔하고 이를 저장한다. 스캔한 정도는 아이폰 X의 A11 바이오닉 칩에 내장된 듀얼 코어 뉴럴 엔진이 등록된 얼굴 정보와 비교한다. 

[ETRI의 얼굴인식	기술개발	동향 보고서(2019)](https://ettrends.etri.re.kr/ettrends/176/0905176004/34-2_29-39.pdf )를 살펴보면 
>“얼굴인식 분야는 전체적으로는 정체기에 들어선 것으로 판단되지만 세부기술 중 얼굴 분류 분야는 성장기에 있으며 감정인식 분야는 기술의 태동기에 있는 것으로 판단되었다. 응용 기술 개발 관점에서는 얼굴분류 분야가 적합할 수 있지만 기초기술 개발을 위해서는 감정인식 분야에 관심을 추천한다. ”

라고 이야기 하고 있다.

## **4. 자율주행 분야**

자율주행의 경우 우리에게 가장 익숙한 제품은 자율주행 자동차일 것이다. 자율주행 자동차는 어렸을 때 미래모습 상상화를 그릴 때 거의 필수로 등장했던 만큼 연구된 역사도 꽤 길다. 실제로 처음 자율주행차가 나온 것은 **1986년 미국 카네기멜론 대학 자율주행 연구팀 내브랩(NavLab)이 밴을 개조하여 만든 ‘내브랩 1’**이다. 이후 많은 자동차회사와 연구기관이 자율주행 자동차를 개발하고 있고 인공지능이 발달하면서 IT업계까지 자율주행차를 선보이고 있다.

[작년 2019년 10월 21일의 기사](http://www.aitimes.kr/news/articleView.html?idxno=14500)를 살펴보면 현대자동차와 기아자동차는 **운전자의 주행 성향에 맞는 부분 자율주행을 구현하는 기술인 SCC-ML**을 세계 최초로 개발했다고 한다. SCC-ML은 크루즈 컨트롤 기능에서 스마트 크루즈 컨트롤(SCC)로, 스마트 크루즈 컨트롤에서 다시 SCC-ML에 발전해 온 것이다. 

![인공지능 자동차](https://user-images.githubusercontent.com/60220571/83725611-2c258280-a67d-11ea-88f7-c4d80c848f83.jpg)

기술의 의미를 순서대로 하나씩 살펴보면 크루즈 컨트롤은 고속도로나 막히지 않는 도로에서 엑셀을 밟지 않고도 일정한 속도를 유지 시켜주는 기능이다. 정속주행 또는 장거리 운행에서 유용하게 사용된다. 여기에 더해 스마트 크루즈 컨트롤(SCC)는 앞 차량과의 거리를 일정하게 유지하여 운전자가 설정해둔 속도로 자율주행을 해주는 기능으로 ADAS(첨단 운전자보조 시스템)의 주요 기술 중 하나다. 

하지만 이 기존의 SCC는 주행패턴을 운전자가 직접 설정해야 하기 때문에 조절단계가 세밀하지 않아 운전 성향을 고스란히 반영할 수 없다는 단점이 있다. 이 때문에 SCC가 운전자의 주행 성향과 다를 경우 운전자가 이질감을 느껴 SCC 사용을 꺼리는 경우도 있었다. 

여기서 발전하여 이번에 현대차·기아차가 독자 개발한 **SCC-ML은 AI 기술을 더해 운전자의 주행 성향을 차가 스스로 학습해 SCC 작동 시 운전자와 거의 흡사한 패턴으로 자율주행을 해준다.** 먼저 전방 카메라, 레이더 등의 센서가 다양한 운전상황에서 발생하는 정보를 지속해서 수집해 ADAS의 제어 컴퓨터로 보낸다. 제어 컴퓨터는 입력된 정보로부터 머신러닝 알고리즘을 통해 운전자의 주행습관을 추출해 종합적인 주행 성향을 파악한다, 

현대차·기아차 자율주행개발센터는 SCC-ML은 머신러닝 알고리즘을 적용해 기존 SCC의 사용성을 대폭 개선했다며 자율주행기술 개발에 필수적인 인공지능 분야에서 업계 선두권의 기술력을 확보했다는 데 의미가 있다고 밝혔다.

자율주행자동차가 상용화되면, 교통사고를 크게 줄일 수 있다는 장점 외에도, 운전하던 시간에서 여가활동이나 다른 생산적인 일에도 활용할 수 있고, 이동의 제약이 많은 노인이나 장애인에게 이동의 편리성을 더 쉽게 제공할 수 있다. 따라서 자율주행 기술은 앞으로도 더욱 발전될 것이며 우리 생활에 밀접한 곳에서 만날 수 있는 날이 곧 올 것이다.
